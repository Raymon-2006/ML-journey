```python
import pandas as pd
```

aggregate functions = Reduces a set of values into a single summary value.
                      Used to summarize and analyze data.
                      Often used with the groupby() function

```python
#Basically we are reducing a bunch of values into a single one
df = pd.read_csv("pokemon.csv")
print(df)
```

We have a dataframe that contains lists of all pokemon from Gen 1 kanto to Gen 6 kalos.

Now lets say we want mean of all values from a column

    Default syntax: print(df.mean())

    But this sends error because in our dataframe their contains columns that have string and we can't calculate mean of string.

So we use print(df.mean(numeric_only = True))

    We pass in the kyword argument in df.mean() called 'numeric_only'.
    We are telling pandas to find the mean of those columns that are numeric.


```python
print(df.mean(numeric_only = True))
```

Now what does these data represent?

From gen 1 to gen 2, The average total stat of all pokemons is 435.10.
This means most are way below the standard strong stats of 600 for pseudo legendaries and 680 for box art legendaries.

Most also have really bad atk stats with average only being 79.001. Keep in mind pokemon like Mega Rayqueza have atk stats of 150 {Which is super high}

Similar is the case for sp.atk stats as pokemon like mega mewtwo Y have sp atk stat of 194 which is highest for all legendaries.

```python
#                   Another aggregate function is the sum()
print(df.sum(numeric_only=True))
#                   This sums all the numeric values of each pokemon in a column
```

Most of the sum data from above is useless Except one which is:

        Their are 65 legendary from pokemon generation 1 to generation 6.

```python
print(df.min(numeric_only=True))            #This returns the smallest value from each  column.
```

```python
print(df.max(numeric_only = True))            #This returns the highest value from each column.
```

```python
print(df.count())                               #This counts the number of rows from each column.
#                                                 count won't include any null value
```

The above aggregate functions apply to the whole data frame. Below is the way aggregate functions work to single column.

```python
print(df["Attack"].mean())                  # We Simply use a subscript operator '[]' on dataframe df to specify index.
```

```python
print(df["HP"].max())                       # Maximun HP in the pokedex
```

```python
print(df["HP"].min())                       # Minimum HP in the pokedex
```

```python
print(df["Name"].count())                   # All pokemon in the dataframe.
#                                              This tell us their are 800 distinct pokemon including 
#                                                mega evolutions the df.
```

Now we can also group together different rows (that have something in common) in the dataframes. Like we can group together pokemon by their types. {.groupby()}

```python
# Lets put grass type in one group, fire type in another and soon.
#       For this we use the groupby() function.

group = df.groupby("Type 1")
print(group)            #This returns a series for us to work with and display.
#                         Remember group is an object.
```

```python
print(group["Attack"].mean())       #We called a aggregate function mean() in our group. We also selected 
#                                     the columne "Attack" for mean value. This returns mean  of attack stat of 
#                                       all the pokemon that have same type 1.

# Judging by the data Dragon type has the highest atk average. with ground at 2nd.
```

```python
print(group["Name"].count())            #For listing no. of pokemon based on type 1, we can also use
#                                         other columne name such as attack, Sp. atk or Sp. Def
```

we can also use .sum(), .min(), .max() values as well for the group.

So these are aggregate functions in pandas. They are used to summarize data.
