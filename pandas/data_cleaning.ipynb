Date cleaning = The process of fixing/removing:
                incomplete, incorrect, or irrelevent data.
                
Around 75% of work done with Pandas is data cleaning.

```python
import pandas as pd
```

```python
df = pd.read_csv("pokemon.csv")
```

The methods are:
1. Drop irrelavent columns (Used of the .drop() function.)

```python
df = df.drop(columns= ["Generation"])
#                               As a keyword argument, We pass in the list of columns we want to drop in the .drop()
#                                The above use of .drop() returns a new dataframe so we reassign it to exiting or a assign
#                                 it to a new dataframe.
print(df)
```

2. Handeling the missing data:

In our dataframe of pokemon.csv, their are many pokemon that don't have the 2nd type. but all pokemon do have the 1st type. 

If the row is missing a value in the type 2 we are removing that entire row. We do it via:

    .dropna() function

```python
df_demo = df.dropna(subset= ["Type 2"])
print(df_demo.head(151))
```

```python
df = df.fillna({"Type 2": "None"})        # Instead of dropping the rows whose values are not available, we can
#                          fill any not available values.
#                           within .filna() we pass in a dictionary.
print(df.head(101))
```

This is how we handle missing data.

3. Fix inconsistent values:

We can replace a instance of a value with another one.

```python
#               Within  a 'type 1' lets select any grass type, we will select instance of grass to 
#                be all upper case letters
df["Type 1"] = df["Type 1"].replace({"Grass": "GRASS"})
print(df.head(51))
```

```python
#               Similarly we can replace 'fire' with 'FIRE' and 'water' with 'WATER'
df["Type 1"] = df["Type 1"].replace({"Grass": "GRASS",
                                     "Water": "WATER",
                                     "Fire": "FIRE"
                                     })
print(df.head(51))
```

 4. Standarize text:
    Changing the case of elements of each columnes.

```python
#           Lets take all the names of pokemon and make them lower case.
df["Name"] = df["Name"].str.lower()
print(df.head(31))
```

 5. Fix data type:

```python
#           In our pokemon.csv, Legendary have 1 and non legendary have 0, I think boolean would make more sense to
#            to use (i.e replacing 1 and 0's with True and False)
df["Legendary"] = df["Legendary"].astype(bool)
print(df.head(160))
```

 6. Removing duplicate values:
 
 If we have duplicates in the .csv files. Like Same pokemon name, atk stats and such in multiple rows, we can remove it via the function:

    .drop_duplicates()

Be sure to assign it to a existant or a new df after using it. 

    df = df.drop_duplicates()
