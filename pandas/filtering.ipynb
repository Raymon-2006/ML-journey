```python
import pandas as pd
```

```python
pokedex = pd.read_csv("pokemon.csv")
```

Filtering = Keeping the rows that match a condition.

```python
# lets create a new df consisting of Pokemon with attack stat of greater then 180
atk_pokemon = pokedex[pokedex["Attack"] >= 180]         #[] is called the subscript operator
print(atk_pokemon)
```

```python
bulky_pokemon = pokedex[pokedex["Defense"]>160] 
print(bulky_pokemon)
```

```python
legendary_pokm = pokedex[pokedex["Legendary"] == True]     #You can also use True instead of 1
print(legendary_pokm)
```

Now what about pokemon with a water typing? Because a pokemon can have water type in both type 1 and type 2. We need to consider if one of two type meet the condition, a pokemon qualifies as a water type.

Here we use 'or' operator 

Note: Here we use the c style or operator | and make sure you don't use 'or' like in python.

```python
water_pokm = pokedex[(pokedex["Type 1"] == "Water") |
                     (pokedex["Type 2"] == "Water")]
print(water_pokm.to_string())
```

```python
print(water_pokm.loc[149])      #Omanyte is a rock/water type and since it appears in the 
#                                water_pokm dataframe, This process is correct
```

Use of and logical operator:
Make sure to use C style and operator (i.e &) and not python 'and'

```python
ff_pokemon = pokedex[(pokedex["Type 1"] == "Fire") & (pokedex["Type 2"] == "Flying")]
#               ff_pokedex refers to all fire and flying pokemon
print(ff_pokemon)
```
