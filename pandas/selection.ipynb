SELECTION BY ROW/S

```python
print(df.loc[0])  #The 0 is a by default assignment by pandas to act as a label.
#                   similar to array index it starts with 0, and in this case prints info about bulbasaur.
```

Now here arises a problem:

You aren't going to remember all 800 pokemon by index from 0 to 799. So we assign a columne to act as a index and in this case we want the name of pokemon to act as index, So we can access all the data regarding a particular pokemon via its name.

```python
pokedex = pd.read_csv("pokemon.csv", index_col="Name")
print(df.head(26))
```

```python
print(pokedex.loc["Bulbasaur"])  #.loc means location by index
#                            .iloc or integer location is position based(integer index)
```

```python
print(pokedex.loc["Dialga"])
```

```python
print(pokedex.loc["Charizard"])
```

```python
# If we don't want all the columne in the specific row, we can 
#  send in a list of columnes we would like to see, alongside the index in
#   the .loc function.
print(pokedex.loc["Zekrom", ["Total", "Attack", "Speed"]])
```

```python
#SELECTION OF PARTICULAR RANGE OF ROW:
#                           We do this via using a slice operator ':'
print(pokedex.loc["Bulbasaur":"Charizard", ["Total", "Attack", "Speed"]])
```

```python
print(pokedex.iloc[0:11])    #slicing to display the first 10 pokemon.
```

```python
#If we need every ODD no. of rows within a partcular set of rows from 1 to 10
print(pokedex.iloc[0:11:2])
```

```python
#Similarly for ODD no. of rows
print(pokedex.iloc[1:11:2])
```

```python
#For second argument in iloc we can select indices of which columns we would like:
print(pokedex.iloc[0:11:2, 0:3])    #I would like to print first 3 columne
```
